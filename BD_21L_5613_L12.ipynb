```python
import pyspark
from pyspark.sql import SparkSession
from pyspark import SparkContext,SparkConf
conf=SparkConf().setAppName('SparkApp').setMaster('local[*]')
sc=SparkContext(conf=conf)
```

# **Question#1**

```python
# spark = SparkSession.builder.getOrCreate()
# df=spark.read.csv("Movies.csv.xlsx")
# df.show()
rdd = sc.textFile("Movies.csv")
rdd1 = rdd.map(lambda x: x.split(","))
filter_rdd = rdd1.filter(lambda x: (x[3] == 'Action' and x[8] == 'Yes'))
result_rdd = filter_rdd.flatMap(lambda x: (x[0],(x[2],x[6])))
result_rdd.collect()
```

# **Question#2**

```python
file_rdd = sc.textFile("Movies.csv")
rdd_split = file_rdd.map(lambda x: x.split(","))
filter_rdd = rdd_split.filter(lambda x: (x[8] == 'Yes'))
result_rdd = filter_rdd.map(lambda x: (x[4],(x[2],x[6])))
result_rdd.reduceByKey(lambda x,y: x+y).collect()
# result_rdd.collect()
```

# Question#**3**

```python
file_rdd = sc.textFile("Movies.csv")
rdd_split = file_rdd.map(lambda x: x.split(","))
filter_rdd = rdd_split.filter(lambda x: (x[8] == 'No'))
result_rdd = filter_rdd.sortBy(lambda x: x[7], ascending=False)
result_rdd.map(lambda x: x[2]).take(10)
# result_rdd.take(10)
```

# **Question#4**

```python
file_rdd = sc.textFile("Movies.csv")
rdd_split = file_rdd.map(lambda x: x.split(","))
filter_rdd = rdd_split.filter(lambda x: (x[0].isdigit()))
result_rdd = filter_rdd.filter(lambda x: int(x[0]) < 1980)
result_rdd.map(lambda x: x[2]).take(10)
```

# Question#**5**

```python
file_rdd = sc.textFile("Movies.csv")
rdd_split = file_rdd.map(lambda x: x.split(","))
filter_rdd = rdd_split.filter(lambda x: (x[0].isdigit()))
rdd_1990 = filter_rdd.filter(lambda x: int(x[0]) < 1990)
result_rdd = rdd_1990.sortBy(lambda x: x[2][0])
result_rdd.map(lambda x: x[2]).collect()
```
